```python
import pandas as pd
import numpy as np
from pydeseq2.dds import DeseqDataSet
from pydeseq2.ds import DeseqStats
```

# Clustered Data

```python
path = "./Kumar/Raw"
marker_path = "./Kumar/Marker"
figures_path = "./Kumar/Figures"
```

```python
expression_matrix = pd.read_csv(path+"/count_table.csv", index_col=0).astype('int')
expression_matrix.columns = [cell.split('.')[0] for cell in expression_matrix.columns]
expression_matrix = expression_matrix[expression_matrix.sum(axis = 1) > 0]
# Load metadata (samples as index, condition as column)
metadata = pd.read_csv(path+"/metadata.csv", index_col=0)
metadata.set_index('Sample', inplace=True)
expression_matrix = expression_matrix.T

# for col in expression_matrix.columns:
    # expression_matrix.loc[expression_matrix[col] < 0, col] = 0

# for col in expression_matrix.select_dtypes(include='number').columns:
#     expression_matrix[col] = expression_matrix[col].clip(lower=0)

# numeric_cols = expression_matrix.select_dtypes(include='number').columns
# expression_matrix[numeric_cols] = expression_matrix[numeric_cols].clip(lower=0)
```

```python
# metadata.index = metadata["Sample"]
```

```python
expression_matrix
```

```python
dds = DeseqDataSet(
    counts=expression_matrix, 
    metadata=metadata,  
    design_factors=["Condition"]
)
```

```python
dds.var_names_make_unique()
```

```python
duplicates = dds.obs_names[dds.obs_names.duplicated()]
print(f"Number of duplicate sample names: {len(duplicates)}")
```

```python
duplicates = dds.var_names[dds.var_names.duplicated()]
print(f"Number of duplicate gene names: {len(duplicates)}")
```

```python
dds.deseq2()
```

```python
stat_res = DeseqStats(dds)
stat_res.summary()
stat_res
```

```python
results_df = stat_res.results_df
```

```python
significant_genes = results_df[
    (results_df["padj"] < 0.05) & (np.abs(results_df["log2FoldChange"]) > 0.5)
]
```

```python
significant_genes.head()
```

```python
sort_sg = significant_genes.dropna().sort_values('padj', ascending=False)
```

```python
sort_sg
```

```python
sort_sg.to_csv(path+"/padj_sort_sg.csv")
```

```python
pos_sg = sort_sg[sort_sg['log2FoldChange'] >= 0]
neg_sg = sort_sg[sort_sg['log2FoldChange'] < 0]
```

```python
pos_sg.to_csv(path+"/pos_sg.csv")
neg_sg.to_csv(path+"/neg_sg.csv")
```

```python
pos_sg
```

```python
neg_sg
```

```python
significant_genes.to_csv(path+"/DEG_results.csv")
```

```python
ranking = pd.DataFrame(zip(significant_genes.index, significant_genes['stat']), columns=['Genes','stat']).dropna().sort_values('stat', ascending=False)
```

```python
ranking
```

```python
gene_list = ranking['Genes'][:99].tolist()
```

```python
genes = np.array(gene_list)
np.savetxt(path+"/topGenes.txt", genes, fmt="%s")
```
